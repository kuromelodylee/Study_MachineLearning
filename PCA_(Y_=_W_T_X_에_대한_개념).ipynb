#####전제: 데이터집합 X는 다차원 벡터 집합이다. (즉, 일반적으로 컬럼값이 인덱스(차원)이 아닌 각행이 차원으로 이루어진 데이터 집합

일반적 데이터 모양
$\begin{pmatrix}x_1&y_1&z_1\\3&4&5\\ \end{pmatrix}$ 

수학에서의 데이터집합
$\begin{pmatrix}x_1&3\\ y_1&4\\ z_1&5\\\end{pmatrix}$ 

고로 수학적으로 다차원의 데이터 행렬은 일반적인 데이터구조와 다르다는 것을 인지할 필요가있다.





### 1. PCA (Y = W.T * X 에 대한 개념)
1. 데이터집합의 정의: 주어진데이터집합 X가 (2000(차원) * 150(테이터수)) 경우

2. 공분산 추출: 공분산을 구하면 차원기준(2000)으로 정방 행렬 2000 * 2000 생성됨

3. 공분산을 통한 고유치 및 고유백터 추출: 위의 공분산 행렬 (2000 * 2000)을 통하여 고유치 대각행렬(2000 * 2000), 고유 열백터 행렬(2000 * 2000)을 구할수 있음
 - np.linalg.eig 리턴값 고유치, 고유백터 (순서대로)
 - 파이썬에서는 대각행렬로 나오지는 않고 대각행렬 대각 요소로 출력됨
 - 대각행렬로 구하는 매써드가 있는데, 요소값이 매우약간 상이함. (np.diag(np.diag(S)))
4. 변환행렬W의 정의: 고유치 대각행렬(2000 * 2000) 대각요소n개와 대응하는 고유 열백터 행렬(2000(차원) * n(개))을 W로 한다.

5. 특징데이터 Y의 정의: 주어진 데이터 집합 X에 대해 변환행렬W의 전치행렬을 곱해 추출한다.
 - 결국 Y = W.T * X 하면 특징데이터는 n개의 차원으로  n * 150크기의 행렬로 특징데이터를 생성한다. (2000 * n).T @ (X (2000 * 150))
  - 일반적으로 데이터를 사용할때는 n이 차원임을 인지하고 150 * n의 데이터 형태로 사용해야 이해하기 쉬움 




###2. 데이터가 위와 같이 매우 큰경우 PCA 방법 (얼굴인식등)
1. 매우크다는 정의: 위와 같이 열이 2000차원을 가진다면 차원기준으로 공분산을 가져 계산량이 많아짐

2. 목적: 공분산 차원 축소후 고유치행렬 및 고유백터행렬을 통해 변환행렬

3. 데이터 집합의 정의: 주어진데이터집합 X가 ((차원)2000 * 150) 경우

4. 축소된 공분산 추출: 주어진데이터 집합 X를 전치((차원)150 * 2000) 시켜서 공분산을 구한다. (150 * 150)크기 공분산 생성
 -  공분산은 일반적인 데이터 형태에서 열벡터(변수, 컬럼등)수로 크기가 결정되는데, (왜냐하면 변수간 상관계를 표현해야하니, 변수(컬럼등) 크기만큼 정방행렬로 만들어진다. 위의 경우에 전치를 한다는 의미는 데이터 집합의 수를 컬럼으로 받겠다는 의미이다. 

5. 공분산을 통한 고유치 및 고유백터 추출: 위의 공분산 행렬 (150 * 150)을 통해 고유치 대각행렬(150 * 150), 고유 열백터 행렬(150 * 150)을 구할수 있음

6. 변환행렬W의 정의
 - (X'-Mean(X').T (크기 2000 * 150)와 위에서 구한 고유열백터행렬 (150 * 150)을 곱하여 변환행렬 W 집합 (2000 * 150)을 추출한다.
    - X(2000 * 150) 대신 X'(150 * 2000)을 사용하고 X'의 평균(1(150) * 2000)을 사용함. 
 - 고유치 대각행렬(150 * 150) 대각요소n개와 대응하는 변환행렬 행렬W의 집합(2000 * n)을 W로 한다.
 - 상기 변환행렬 W 집합의 의미: 변환행렬의 고유백터의 개수는 150개이며, 차원은 2000 차원 (X의 학습데이터 개수만큼 고유백터가 개수가 의존적으로됨) 1개의의 고유치에 대응하는 고유열 벡터는 결국 2000개의 데이터(차원)를 가진 1개의 벡터를 의미한다.

5. 특징데이터 Y의 정의: 주어진 데이터 집합 X와 변환행렬 W의 전치행렬을 곱해 추출한다.
 - 결국 Y = W.T * X 하면 특징데이터는 n개의 차원으로  n * 150 크기의 행렬로 특징데이터를 생성한다. (2000 * n).T @ (X(2000 * 150)) 
 이번의 경우는 고유값들을 추출 위해서 X의 전치를 활용하는 것이 아니므로  X는 기존 모양그대로 사용한다.
 - 일반적으로 데이터를 사용할때는 n이 차원임을 인지하고 150 * n의 데이터 형태로 사용해야 이해하기 쉬움 



실매트랩과 파이썬의 경우 데이터 (150 * 2000(차원))로 공분산을 구하면
 - 매트랩의 경우 컬럼을 기준으로 공분산의 크기를 결정 (2000 * 2000)
 - 파이썬 넘파이의 경우는 default가 로우를 기준으로 공분산의 크기를 결정 (150 * 150)

즉, 파이썬은 수학적으로 데이터의 행렬을 인지하고 매트랩은 데이터 세트를 practical하게 인지한다.

numpy:
